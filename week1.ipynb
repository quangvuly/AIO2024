```python
# Câu 1: Viết function thực hiện đánh giá classification model bằng F1-Score.
def is_number(n):
  if type(n) == int:
    return True
  return False

def is_larger_0(n):
  if n > 0:
    return True
  return False

def calc_precision(tp, fp):
    return tp / (tp + fp)

def calc_recall(tp, fn):
    return tp / (tp + fn)

def calc_f1_score(tp, fp, fn):
    precision = calc_precision(tp, fp)
    recall = calc_recall(tp, fn)
    return 2 * (precision * recall) / (precision + recall)

def exercise1(tp, fp, fn):
  if not is_number(tp):
    raise ValueError('tp must be int')
  if not is_number(fp):
    raise ValueError('fp must be int')
  if not is_number(fn):
    raise ValueError('fn must be int')
  if not is_larger_0(tp) and not is_larger_0(fp) and not is_larger_0(fn):
    raise ValueError('tp and fp and fn must be greater than zero')

  return calc_f1_score(tp, fp, fn)

print(exercise1(2, 3, 0))
```

```python
# Câu 2: Viết function mô phỏng theo 3 activation function.

def is_number(n):
  try:
    float(n)
  except ValueError:
    return False
  return True

def is_correct_activation(activation):
  if activation == 'sigmoid' or activation == 'relu' or activation == 'elu':
    return True
  return False

def calc_sigmoid(x):
  return 1/(1 + math.exp(-x))

def calc_relu(x):
  if x > 0:
    return x
  return 0

def calc_elu(a,x):
  if x > 0:
    return x
  return a * (math.exp(x) -1 )

def exercise2():
  x = input('Input x: ')
  if not is_number(x):
    raise ValueError('x must be a number')
  else:
    x = float(x)

  activation = input('Input activate function (sigmoid|relu|elu): ')
  if not is_correct_activation(activation):
    print (f' {activation} is not supported')

  if activation == 'sigmoid':
    return calc_sigmoid(x)
  if activation == 'relu':
    return calc_relu(x)
  if activation == 'elu':
    print ('Default a = 0.01')
    return calc_elu(0.01, x)

def calc_activation(x,activation):
  if activation == 'sigmoid':
    return calc_sigmoid(x)
  if activation == 'relu':
    return calc_relu(x)
  if activation == 'elu':
    print ('Default a = 0.01')
    return calc_elu(0.01, x)

# exercise2()
# print(round(calc_elu(0.01,-1),2))
print(round(calc_activation(3,'sigmoid'),2))
```

```python
# Câu 4 Viết 4 functions để ước lượng các hàm số sau.

def giai_thua(n):
  result = 1
  for i in range(1,n+1,1):
    result *= i
  return result

def calc_sin(x,n):
  result = 0
  for i in range(n):
    result += pow(-1,i)* pow(x,2*i+1) / giai_thua(2*i+1)
  return result

def calc_cos(x,n):
  result = 0
  for i in range(n):
    result += pow(-1,i)* pow(x,2*i) / giai_thua(2*i)
  return result

def calc_sinh(x,n):
  result = 0
  for i in range(n):
    result += pow(x,2*i+1) / giai_thua(2*i+1)
  return result

def calc_cosh(x,n):
  result = 0
  for i in range(n):
    result += pow(x,2*i) / giai_thua(2*i)
  return result
print (round(calc_sin(3.14,10),4))
print (round(calc_cos(3.14,10),2))
print (round(calc_sinh(3.14,10),4))
print (round(calc_cosh(3.14,10),4))
```

```python
# Câu 5

def x_root(a,x):
  return a ** (1 / x)

def md_nre_single_sample (y, y_hat, n, p):
  return pow((x_root(y,n) - x_root(y_hat,n)),p)

print(md_nre_single_sample (y=100, y_hat=99.5, n=2 ,p=1))
print(md_nre_single_sample (y=50, y_hat=49.5, n=2 ,p=1))
print(md_nre_single_sample (y=20, y_hat=19.5, n=2 ,p=1))
print(md_nre_single_sample (y=0.6, y_hat=0.1, n=2 ,p=1))
```
